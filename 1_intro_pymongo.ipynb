Selecione o documento que possui o maior número de solicitações QtdeSolic na Anatel.

```python
from bson.son import SON
import pprint
parametro = [ {"$group": { "_id": "null", "maxQuantity": {"$max": "$QtdeSolic"}  } }]
client.db.minas.find_one({'QtdeSolic': 'parametro'})
list(db.minas.find({'QtdeSolic': 4164}))
```

Qual é o menor número de solicitações QtdeSolic na Anatel?

```python
from bson.son import SON
import pprint
parametro = [ {"$group": { "_id": "null", "maxQuantity": {"$min": "$QtdeSolic"}  } }]
pprint.pprint(list(db.minas.aggregate(parametro)))
```

Mostre todos os documentos com o menor número de solicitações QtdeSolic na Anatel.

```python
list(db.minas.find({'QtdeSolic': '1'}))
```

Mostre todos os documentos com atributo Tipo igual a "Reclamação".
